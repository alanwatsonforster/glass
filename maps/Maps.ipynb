## Inverted Map Sheets

Inverted map sheets are used in:

- V-11: B2
- V-22: C1 and C2

Map sheets can be inverted by appending `/i`. For example, this shows the map for scenario V-11 in TSOH:

```python
import sys
sys.path.append('../src')
from glass import *
import os

dotsperhex = 40

startgamesetup(
    None,
    sheets=[["A1"], ["B2"], ["B1"]],
    invertedsheets=["B2"],
    style="airstrike",
    dotsperhex=dotsperhex,
    printlog=False,
    writelogfiles=False,
)
drawmap(zoom=False)
os.rename("map-00.png", "map-V-11.png")
```
